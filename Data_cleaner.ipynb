```python
import numpy as np
import pandas as pd
import os
```

## Concate all data file inside specified directory

```python
def concate_all_files(path):
    dfs = []
    for file in os.listdir(path):
        if file.endswith(".pkl"):
            dfs.append( pd.read_pickle(path+file) )
        elif file.endswith(".csv"):
            if not(pd.read_csv(path+file).empty):
                dfs.append( pd.read_csv(path+file) )
            
    print("found", len(dfs))
    
    print("proceed to concatenation ...")
    all_dfs = pd.concat(dfs)
    print("... concatenation finished:", all_dfs.shape)
    
    return all_dfs
```

```python
concated_df = concate_all_files("Data-to-clean/")
```

## Delete duplicated row based on Listing Number (unique)

```python
def delete_duplicated_row(df):
    old_len = len(df)
    df = df.drop_duplicates(subset='Listing Number', keep="first")
    new_len = len(df)
    print('>> deleted', old_len-new_len, 'rows')
    
    return df
```

```python
concated_df = delete_duplicated_row(concated_df)
```

```python
concated_df.shape
```

## Delete empty rows (all features are NaN)

```python
def delete_NaN(df):
    index_names = df[ df['Listing Number'].isnull() ].index
    
    old_len = len(df)
    df = df.drop(index_names, inplace = False)
    new_len = len(df)
    print('>> dropped', old_len-new_len, 'empty rows')
    
    return df
```
