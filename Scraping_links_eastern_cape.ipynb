```python
import requests
import numpy as np
from bs4 import BeautifulSoup
import re
import dateutil
from random import choice
```

```python
def proxy_generator():
    response = requests.get("https://sslproxies.org/")
    soup = BeautifulSoup(response.content, 'html5lib')
    proxy = {'https': choice(list(map(lambda x:x[0]+':'+x[1], list(zip(map(lambda x:x.text, soup.findAll('td')[::8]), map(lambda x:x.text, soup.findAll('td')[1::8]))))))}
    
    return proxy

def data_scraper(request_method, url, **kwargs):
    while True:
        try:
            proxy = proxy_generator()
            print("Proxy currently being used: {}".format(proxy))
            response = requests.request(request_method, url, proxies=proxy, timeout=7, **kwargs)
            break
            # if the request is successful, no exception is raised
        except:
            print("Connection error, looking for another proxy")
            pass
    return response
```

```python
def get_pages(page_url, nb):
    pages = []
    for i in range(1, nb+1):
        j = page_url + "p" + str(i)
        pages.append(j)
    return pages
```

```python
def get_links(url, page_url, town):
    links = []

    print("--- Getting max pagination:")
    
    response = data_scraper('get', page_url)
    src = response.content
    document = BeautifulSoup(src, 'lxml')
    max_pagination = document.find("ul", {"class": "pagination"}).find_all("li")[-1].find("a").text
    print (max_pagination)

    pages = get_pages(page_url, int(max_pagination))

    for page in pages:
        
        print(page)
        
        response = data_scraper('get', page)
        src = response.content
        document = BeautifulSoup(src, 'lxml')

        js_listingResultsContainer = document.find_all("div", {"class":"js_listingResultsContainer"})
        
        print("js_listingResultsContainer size: ")
        print(len(js_listingResultsContainer))
        
        if len(js_listingResultsContainer) != 0:
            js_resultTiles = js_listingResultsContainer[-1].find_all("div", {"class": "js_resultTile"})
        else:
            js_resultTiles = []

        for js_resultTile in js_resultTiles:
            a_1 = js_resultTile.find("a")

            links.append(url + a_1.attrs['href'])
            
            print(len(links))
            
            np.save("links_from_" + town, np.array(links))

    return links
```

```python
# scraping links from eastern_cape province

page_url = 'https://www.property24.com/to-rent/eastern-cape/7/'
url = 'https://www.property24.com'

get_links(url, page_url, "easter_cape")
```
